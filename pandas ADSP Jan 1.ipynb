## Series
A Series in pandas is a One dimensional array like structute with data and its label/index. in addition to the data values Series also display and allows to modify the index of the data.
A Series can be created from a list, array, dictionary, scalars containing integers, floating, string, any python object

```python
import numpy as np
import pandas as pd
```

```python
w = [3,1,2,0,5]
```

```python
wr =  np.array(w)
```

```python
print(wr)
```

```python
# Creating pandas series from list with default index
wrs =  pd.Series(w)
print(wrs)
```

```python
# Creating pandas series from list with user defined index
wrs =  pd.Series(w, index = ['Shirts','Trousers', 'Wallets', 'Money', 'Shoes'])
print(wrs)
```

## DataFrame
A DataFrame in pandas is a two dimensional array like structute with data and its labels for rows and columns. in addition to the data values Series also display and allows to modify the row indices and columns labels of the data.
A DataFrame can be created from a list, array, dictionary, scalars, Series containing integers, floating, string, any python object

### Creating DataFrames
There are two commonly used ways to create DataFrames from pandas
- using DataFrame() to create from lists/arays/dictionaries
- reading data from external files like csv, excel, sql, json etc

#### generic syntax for dataframe creation using DataFrame ()
pd.DataFrame(`data', index = `list of indices`, columns = `list of colmn names`)

```python
## Creating DataFrame from Dictionary
dict1 =  {
    'City': ['London', 'Sydney', 'Dubai', 'Mumbai', 'Johannesberg', 'Colombo'],
    'Year' :[1995, 1999, 2003, 2007, 2013, 2018],
    'Winner':['IND', 'SA', 'SL', 'AUS', 'IRE', 'ZIM']
}
```

```python
exp =  pd.Series(dict1)
print(exp)
```

```python
df1 =  pd.DataFrame(dict1)
print(df1)
```
